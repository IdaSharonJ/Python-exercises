```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#setting ignore as parameter
import warnings
warnings.filterwarnings('ignore') 
```

```python
df = pd.read_csv('Cars.csv')
df.head()
```

```python
df.info()
```

```python
df.drop('sno', axis=1, inplace=True)
df.drop('rev_permile', axis=1, inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.hist(layout=(3,3),figsize=(20,15))
plt.show()
```

```python
sns.distplot(df['minprice'],kde=True,label='True')
plt.axvline(x=df['minprice'].median(),color='green',ls='--')
plt.axvline(x=df['minprice'].mean(),color='red',ls='--')
plt.show()
```

The Variable 'minprice' does not follow normal distribution.

```python
sns.distplot(df['price'],kde=True,label='True')
plt.axvline(x=df['price'].median(),color='green',ls='--')
plt.axvline(x=df['price'].mean(),color='red',ls='--')
plt.show()
```

The variable 'price' does not follow normal distribution.
