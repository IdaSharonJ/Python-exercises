```python
import numpy as np
import pandas as pd
import datetime
```

**DATA** **FRAMES**

```python
students = ['Naresh', 'Ajay', 'Harish', 'Yusuf']
```

```python
scores = [87, 83, 93, 82]
```

```python
exam_results = pd.DataFrame(
    {
        "name": students,
        "scores": scores
    }
)
```

```python
exam_results
```

**SEARCHING**

```python
exam_results.loc[0, 'scores']
```

```python
exam_results.loc[:, 'scores']
```

```python
exam_results.loc[2, :]
```

**ADD COUMNS**

```python
age = [24,23,20,22]
gender = ['male','male','male','male']
```

```python
exam_results['age'] = age
exam_results['gender'] = gender
exam_results
```

```python
exam_results.drop('gender', axis=1)
```

**GROUP BY**

```python
exam_results.groupby('gender').mean()
```

```python
mylist = [1,2,1,2]

exam_results.groupby(mylist).get_group(1)
```

```python
exam_results.groupby(mylist).get_group(2)
```

```python
above_85 = exam_results['scores'] > 85
above_85
```

```python
exam_results.groupby(above_85).mean()
```

**MERGING AND JOINING**

```python
contacts = pd.DataFrame(
    {
        'name' : ['Harish', 'Naresh'],
        'tel'  : ['+44 1234 5678', '+44 3245 5564']
    }
)
```

```python
exam_results.merge(contacts, left_on='name', right_on='name')  
```

```python
exam_results.merge(contacts, left_on='name', right_on='name', how='outer')
```
