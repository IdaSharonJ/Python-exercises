```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data =pd.read_csv('uscrimes.csv') 
data.head(3)
```

```python
data.describe()
```

```python
data.columns
```

```python
data.isna
```

```python
sns.distplot(data['Year'],kde=False)
```

```python
sns.jointplot(x='Month',y='Crime Name1',data=data,color='red')
```

```python
sns.rugplot(data['Month'],color='green')
```

```python
sns.kdeplot(data['Year'])
```

```python
sns.barplot(x='Crime Name1',y='Month',data=data,color='yellow')
```

```python
sns.boxplot(y='Month',x='Victims',data=data,palette='jet_r')
```

```python
sns.violinplot(x='Crime Name1',y='Year',data=data)
```

```python
sns.stripplot(x='Month',y='Victims',data=data,palette='GnBu')
```

```python
sns.displot(data=data, x="Month", hue="Police District Number", kind="kde", fill=True)
```

```python
cordata=data.corr()
cordata
```

```python
sns.heatmap(cordata,annot=True,cmap='crest')
```
